```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
maindata['AnimalName']=le.fit_transform(maindata['AnimalName'])
maindata['symptoms1']=le.fit_transform(maindata['symptoms1'])
maindata['symptoms2']=le.fit_transform(maindata['symptoms2'])
maindata['symptoms3']=le.fit_transform(maindata['symptoms3'])
maindata['symptoms4']=le.fit_transform(maindata['symptoms4'])
maindata['symptoms5']=le.fit_transform(maindata['symptoms5'])
maindata['Dangerous']=le.fit_transform(maindata['Dangerous'])


```

```python
maindata.head()
```

```python
target='Dangerous'
independent=[x for x in maindata.columns if x!=target]
x=maindata[independent]
y=maindata[target]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Definging all the models


ModelRF=RandomForestClassifier()
ModelET = ExtraTreesClassifier()
ModelDT=DecisionTreeClassifier()
ModelLR=LogisticRegression()
ModelKNN = KNeighborsClassifier(n_neighbors=5)
ModelSVM = SVC(probability=True)
ModelSVM = GaussianNB(priors=None, var_smoothing=1e-09)

models=[ModelRF,ModelET,ModelDT,ModelLR,ModelKNN,ModelSVM,ModelSVM]
model_names=["Random Forest","Extra Trees Classifier","Decision Tree","Logistic Regression","Neighbor Classifier","SVC","GuassianNB"]
```

```python
def RUN_ALGORITHMS():
  result={
      "Model Name":[],
      "tp":[],
      "fn":[],
      "fp":[],
      "tn":[],
      "accuracy":[],
      "precision":[],
      "specificity":[],
      "recall":[],
      "f1 Score":[],
      "roc_auc_score":[],
      "balanced_accuracy":[],
      "MCC":[]
  }
  result=pd.DataFrame(result)
  for i in range(len(models)):
    Model=models[i]
    #testing with x_test into y_pred
    Model = Model.fit(x_train,y_train)
    y_pred = Model.predict(x_test).astype(int)
    actual = y_test
    predicted = y_pred
   #Generating CONFUSION MATRIX
    matrix = confusion_matrix(actual,predicted,labels=[1,0],sample_weight=None,normalize=None)
    tp,fn,fp,tn=matrix.reshape(-1)
    classification_rep = classification_report(actual,predicted,labels=[1,0])


    #Calculating parameters
    sensitivity = round(tp/(tp+fn), 3);
    specificity = round(tn/(tn+fp), 3);
    accuracy = round((tp+tn)/(tp+fp+tn+fn), 3);
    balanced_accuracy = round((sensitivity+specificity)/2, 3);
    precision = round(tp/(tp+fp), 3);
    f1Score = round((2*tp/(2*tp + fp + fn)), 3);
    from math import sqrt


    #plt.savefig('Log_ROC')
    mx = (tp+fp) * (tp+fn) * (tn+fp) * (tn+fn)
    MCC = round(((tp * tn) - (fp * fn)) / sqrt(mx), 3)
    roc_auc=round(roc_auc_score(actual,predicted),3)
    #printing observations
    result.loc[len(result.index)] = [model_names[i],tp,fn,fp,tn,accuracy*100,precision,specificity,sensitivity,f1Score,roc_auc,balanced_accuracy,MCC]
  result.to_csv("/content/result_analysis.csv")
#models list
RUN_ALGORITHMS()
```
